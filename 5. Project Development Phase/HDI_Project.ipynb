```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import pickle
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv(r"C:\Users\balak\Desktop\HDI_Project\Human Development Index - Full.csv.csv")

print("Dataset loaded successfully!")
print(df.shape)
```

```python
# Select required columns - 2021 data only
df_selected = df[['Country', 
                   'Human Development Groups',
                   'Human Development Index (2021)',
                   'Life Expectancy at Birth (2021)',
                   'Expected Years of Schooling (2021)',
                   'Mean Years of Schooling (2021)',
                   'Gross National Income Per Capita (2021)']].copy()

# Rename columns for easy use
df_selected.columns = ['Country', 'HDI_Category', 'HDI_Score', 
                        'Life_Expectancy', 'Expected_Schooling', 
                        'Mean_Schooling', 'GNI_Per_Capita']

print(df_selected.shape)
print(df_selected.head())
print(df_selected['HDI_Category'].unique())
```

```python
# Cell 3 - Data Visualization (EDA)

# 1. HDI Category Distribution
plt.figure(figsize=(8,5))
df_selected['HDI_Category'].value_counts().plot(kind='bar', color=['red','orange','blue','green'])
plt.title('HDI Category Distribution')
plt.xlabel('HDI Category')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()

# 2. HDI Score Distribution
plt.figure(figsize=(8,5))
sns.histplot(df_selected['HDI_Score'].dropna(), bins=20, kde=True, color='blue')
plt.title('HDI Score Distribution')
plt.xlabel('HDI Score')
plt.show()

# 3. Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df_selected.drop(['Country','HDI_Category'], axis=1).corr(), 
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

print("Visualizations complete!")
```

```python
# Cell 4 - Data Preprocessing

# Drop null values
df_clean = df_selected.dropna()
print("After dropping nulls:", df_clean.shape)

# Label Encoding - HDI Category to numbers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_clean = df_clean.copy()
df_clean['HDI_Label'] = le.fit_transform(df_clean['HDI_Category'])
print("\nHDI Category encoding:")
print(dict(zip(le.classes_, le.transform(le.classes_))))

# Independent variables (X) and Dependent variable (y)
X = df_clean[['Life_Expectancy', 'Expected_Schooling', 
               'Mean_Schooling', 'GNI_Per_Capita']]
y = df_clean['HDI_Score']

print("\nX shape:", X.shape)
print("y shape:", y.shape)
print("\nPreprocessing complete!")
```

```python
# Cell 5 - Train Test Split and Model Training

# Split data 75% train, 25% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

# Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Model Accuracy
r2 = r2_score(y_test, y_pred)
print("\nR2 Score:", round(r2*100, 2), "%")
print("Model Training Complete!")

# Actual vs Predicted Plot
plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], 
         [y_test.min(), y_test.max()], 
         color='red', linewidth=2)
plt.xlabel('Actual HDI Score')
plt.ylabel('Predicted HDI Score')
plt.title('Actual vs Predicted HDI Score')
plt.show()
```

```python
# Cell 6 - Save Model using Pickle

import pickle

# Save model
with open(r"C:\Users\balak\Desktop\HDI_Project\HDI.pkl", 'wb') as f:
    pickle.dump(model, f)

# Save label encoder
with open(r"C:\Users\balak\Desktop\HDI_Project\label_encoder.pkl", 'wb') as f:
    pickle.dump(le, f)

print("Model saved successfully!")
print("HDI.pkl file created!")
```

```python
# Cell 7 - Create Flask App

flask_code = '''
from flask import Flask, request, render_template
import pickle
import numpy as np

app = Flask(__name__)

# Load model
model = pickle.load(open('HDI.pkl', 'rb'))

@app.route('/')
def home():
    return render_template('indexnew.html')

@app.route('/predict', methods=['POST'])
def predict():
    life_expectancy = float(request.form['life_expectancy'])
    expected_schooling = float(request.form['expected_schooling'])
    mean_schooling = float(request.form['mean_schooling'])
    gni_per_capita = float(request.form['gni_per_capita'])
    
    features = np.array([[life_expectancy, expected_schooling, 
                          mean_schooling, gni_per_capita]])
    prediction = model.predict(features)[0]
    
    if prediction >= 0.8:
        category = "Very High"
    elif prediction >= 0.7:
        category = "High"
    elif prediction >= 0.55:
        category = "Medium"
    else:
        category = "Low"
    
    return render_template('resultnew.html', 
                          hdi_score=round(prediction, 3),
                          hdi_category=category)

if __name__ == "__main__":
    app.run(debug=True)
'''

# Save app.py
with open(r"C:\Users\balak\Desktop\HDI_Project\app.py", 'w') as f:
    f.write(flask_code)

print("app.py created!")
```
